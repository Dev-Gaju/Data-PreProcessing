## Question 2

```python
categories = ['comp.graphics','comp.os.ms-windows.misc',
              'comp.sys.ibm.pc.hardware',
              'comp.sys.mac.hardware','rec.autos','rec.motorcycles',
              'rec.sport.baseball','rec.sport.hockey']
train_dataset = fetch_20newsgroups(subset = 'train', categories = categories,
                                   shuffle = True, random_state = 42)  
test_dataset = fetch_20newsgroups(subset = 'test', categories = categories,
                                  shuffle = True, random_state = 42)
```

```python
lemmatizer = WordNetLemmatizer()   #go root of each word 
vectorizer = CountVectorizer(stop_words='english', min_df=3)
# min_df remove word that have count less than 3

tfidf_transformer = TfidfTransformer()   #find more inresting and usefull word

def partsofSpeech(penntag): #reference: discussion notebook
    morphy_tag = {'NN':'n', 'JJ':'a',
                  'VB':'v', 'RB':'r'}
#noun, adjective, verb, adverb
    try:
        return morphy_tag[penntag[:2]]
    except:
        return 'n' 

def lemmatizer_func(sentence): 
    lemmatized_sen = []
    lemma_list = [lemmatizer.lemmatize(word.lower(), pos=partsofSpeech(tag)) 
            for word, tag in pos_tag(word_tokenize(sentence))]
    for lemma in lemma_list:
        if (not any(char in lemma for char in punctuation) and 
            not any(char.isdigit() for char in lemma)):  #remove punctuation
            lemmatized_sen.append(lemma.lower())
    return ' '.join(lemmatized_sen)   
```

```python
# if given input similer then return the word 
partsofSpeech('JJbAbAAA NN VBD dortmund NN') 
```

```python
lemmatizer_func("I played, football")
```

```python
# Now Fit the method to data and check the output
```

```python
train_data_proc = []
test_data_proc = []
for i in range(len(train_dataset.data)):
    train_data_proc.append(lemmatizer_func(train_dataset.data[i]))
for i in range(len(test_dataset.data)):
    test_data_proc.append(lemmatizer_func(test_dataset.data[i]))
    
train_data_feat_vec = vectorizer.fit_transform(train_data_proc)
test_data_feat_vec = vectorizer.transform(test_data_proc)
train_data_feat = tfidf_transformer.fit_transform(train_data_feat_vec)
test_data_feat = tfidf_transformer.transform(test_data_feat_vec)
```

### Question 3
Reduce Dimension from TFIDF vecto

```python
#define two technique
svd = TruncatedSVD(n_components=50, random_state=42) 
nmf = NMF(n_components=50, init='random', random_state=42)
```

```python
#training with LSI and NMF
train_data_LSI = svd.fit_transform(train_data_feat)
print('LSI Train Data Shape:', train_data_LSI.shape)
train_data_NMF = nmf.fit_transform(train_data_feat)
print('NMF Train Data Shape:', train_data_NMF.shape)
U_tr,S_tr,V_tr = randomized_svd(train_data_feat,n_components=50,
                                random_state=42)
print('LSI (train) error:',np.sum(np.array(train_data_feat -
                                           (U_tr.dot(np.diag(S_tr))
                                            .dot(V_tr)))**2))
print('NMF (train) error:',np.sum(np.array(train_data_feat - train_data_NMF.
                                           dot(nmf.components_))**2))

```

```python
#Testing with LSI and NMF
test_data_LSI = svd.transform(test_data_feat)
print('LSI Test Data Shape:', test_data_LSI.shape)
test_data_NMF = nmf.transform(test_data_feat)
print('NMF Test Data Shape:', test_data_NMF.shape)
U_te,S_te,V_te = randomized_svd(test_data_feat,n_components=50,random_state=42)
print('LSI (test) error:',np.sum(np.array(test_data_feat - 
                                    (U_te.dot(np.diag(S_te)).dot(V_te)))**2))
print('NMF (test) error:',np.sum(np.array(test_data_feat - test_data_NMF.
                                          dot(nmf.components_))**2))
```

```python
#Testing with LSI and NMF
test_data_LSI = svd.transform(test_data_feat)
print('LSI Test Data Shape:', test_data_LSI.shape)
test_data_NMF = nmf.transform(test_data_feat)
print('NMF Test Data Shape:', test_data_NMF.shape)
U_te,S_te,V_te = randomized_svd(test_data_feat,n_components=50,random_state=42)
print('LSI (test) error:',np.sum(np.array(test_data_feat - 
                                    (U_te.dot(np.diag(S_te)).dot(V_te)))**2))
print('NMF (test) error:',np.sum(np.array(test_data_feat - test_data_NMF.
                                          dot(nmf.components_))**2))
```

Soft margin always better for classification.
In the case of hard margin(c value increasing) it couldn't classify data properly even classified data. sometimes it pick outliers and give result.

```python
svm_hard = svm.SVC(kernel='linear', C=1000, random_state=42)
svm_soft = svm.SVC(kernel='linear', C=0.0001, random_state=42)
#EXCUTE WITH TRAIN TEST
pred_hard = svm_hard.fit(train_data_LSI,y_train).predict(test_data_LSI)
pred_soft = svm_soft.fit(train_data_LSI, y_train).predict(test_data_LSI)
```
